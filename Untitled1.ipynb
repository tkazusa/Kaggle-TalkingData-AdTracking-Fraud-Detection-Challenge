```python
import pandas as pd
df = pd.read_csv("勤怠まとめ.csv")
df.head()
```

```python
df_time = df["時間"].str.split(':', expand=True).astype('int')
df_time.columns = ['hour', 'min', 'second']

df["日付"] = pd.to_datetime(df["日付"], format='%Y/%m/%d')
df["year"] = df["日付"].dt.year.astype('int')
df["month"] = df["日付"].dt.month.astype('int')

df2 = pd.concat([df[["社員名", "日付", "year", "month", "ジョブコード", "ジョブ名", "時間"]], df_time], axis=1)
```

```python
type(df_time["hour"][0])
```

```python
import numpy as np

a = df2.groupby(["ジョブ名", "社員名"]).agg({"hour": "sum", "min": "sum"})
a["MinToHour"] = a["min"] / 60
a["TotalHour"] = np.round(a["MinToHour"] + a["hour"]).astype('int') 
a["工数"] = a["TotalHour"] / (8*20) 
a
```
